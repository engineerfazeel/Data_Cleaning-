```python
train_metadata = pd.read_csv("D:/Data/train_metadata.csv")
```

```python
train_metadata.head()
```

```python
# Directly drops unnecessary coloumns
train_metadata = train_metadata.drop(columns=['scientific_name', 'common_name', 'author', 'license', 'url'])
# Rename coloumns 
train_metadata = train_metadata.rename(columns={'primary_label': 'primary label'})
train_metadata = train_metadata.rename(columns={'secondary_labels': 'secondary labels'})
train_metadata = train_metadata.rename(columns={'filename':'filenames'})
# A function to remove brackets, quotes and spaces
def remove_brackets_quotes_spaces(string: str) -> str:
    return re.sub(r'[\[\]\' ]', '', string)
# Function usage 
train_metadata['secondary labels'] = train_metadata['secondary labels'].map(remove_brackets_quotes_spaces)
train_metadata['type'] = train_metadata['type'].map(remove_brackets_quotes_spaces)
# Removing only those rows which had empty coloumns 
train_metadata = train_metadata[train_metadata['secondary labels'] != ''].reset_index(drop=True)
train_metadata = train_metadata[train_metadata['time'] != '?'].reset_index(drop=True)
train_metadata = train_metadata[train_metadata['time'] != 'xx:xx'].reset_index(drop=True)
# Saving data frame to a csv file with url
train_metadata.to_csv("D:/Data/train_metadata_cleaned.csv")
```
